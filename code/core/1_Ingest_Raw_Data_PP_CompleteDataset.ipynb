```python
filename = '/Volumes/caio_moreno/uk_house_advisor_bot/files/pp-complete.csv'
print(filename)

# Load the CSV file into a DataFrame
df_silver = spark.read.format("csv").option("header", False).load(filename)

display(df_silver)
```

```python
# Write the DataFrame to a Delta table
df_silver.write.format("delta").mode("overwrite").saveAsTable("caio_moreno.uk_house_advisor_bot.raw_pp_complete")
```

```python
sql_df = spark.sql("SELECT * FROM caio_moreno.uk_house_advisor_bot.raw_pp_complete")
display(sql_df)
```

```python
# Check if the result of the Spark SQL query in sql_df has more than one row
if sql_df.count() > 1:
    print("The query result has more than one row.")
    print("The query result number of rows = " + str(sql_df.count()))
else:
    print("The query result has one or zero rows.")
    print("The query result number of rows = " + str(sql_df.count()))
```
