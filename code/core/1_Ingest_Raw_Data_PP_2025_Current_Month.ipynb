```python
filename = '/Volumes/caio_moreno/uk_house_advisor_bot/files/pp-2025.csv'
print(filename)

# Load the CSV file into a DataFrame
df_raw = spark.read.format("csv").option("header", False).load(filename)

display(df_raw)
```

```python
# Write the DataFrame to a Delta table
df_raw.write.format("delta").mode("overwrite").saveAsTable("caio_moreno.uk_house_advisor_bot.raw_pp_2025")
```

```python
sql_df = spark.sql("SELECT * FROM caio_moreno.uk_house_advisor_bot.raw_pp_2025")
display(sql_df)
```

```python
# Check if the result of the Spark SQL query in sql_df has more than one row
if sql_df.count() > 1:
    print("The query result has more than one row.")
    print("The query result number of rows = " + str(sql_df.count()))
else:
    print("The query result has one or zero rows.")
    print("The query result number of rows = " + str(sql_df.count()))
```
