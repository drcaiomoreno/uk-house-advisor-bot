In the Price Paid Data dataset, "SAON" stands for Secondary Addressable Object Name. It is used to specify sub-building information at an address, such as a flat, apartment, suite, or unit within a larger building. For example, it might include details like "Flat 2," "Apartment B," or "Unit 5B," helping to distinguish individual residences or premises within a single main address.​

%md
In the Price Paid Data dataset, "PAON" stands for Primary Addressable Object Name. It typically refers to the main identifier for an address, such as the house number or the name of the building. For example, "221B" in "221B Baker Street" or "The Laurels" in a rural property. This field helps uniquely identify the specific location of a property within an address

In the context of the Price Paid Data dataset, the "duration" column indicates the term of the title, specifically whether the property is held as freehold or leasehold. For leasehold properties, this field shows the lease duration (for example, 99 years or 125 years), while for freehold properties, the duration is usually marked as "freehold". This information helps distinguish between permanent ownership and properties held under a fixed lease term.

F = Freehold
L = Leasehold

In the Price Paid Data dataset, the "old_new" column indicates whether the property transaction relates to a newly built property or an established (existing) one. The typical values are:​

N: Newly built property​

Y: Existing property (not a new build)​

This distinction helps users analyze price trends and volume differences between new construction and resale properties in the market.

%md
PPD category B differs from category A in that category B refers to "Additional Price Paid entries," which include transactions such as transfers under a power of sale, repossessions, buy-to-lets (where they are identified by a mortgage lender), and transfers to non-private individuals. In contrast, category A includes the standard sales of a single residential property to a private individual and is considered more typical for market comparisons. Category B transactions may reflect special circumstances and are less representative of typical market activity.

The column "record_status" in the Price Paid Data dataset indicates whether a transaction record is current, amended, or has been deleted or replaced. The most common values are:​

A: Addition (new record added to the dataset)​

C: Change (record has been changed or corrected)​

D: Deletion (record has been deleted, often due to error or duplication)​

This column helps users track updates, corrections, and removals within the dataset, ensuring analyses are based on the most accurate and up-to-date information available.
