```python
%sql
SELECT
  DATE_TRUNC('month', CAST(`_c2` AS DATE)) AS sale_month,
  COUNT(*) AS sales_volume
FROM
  `caio_moreno`.`uk_house_advisor_bot`.`raw_pp_complete`
GROUP BY
  sale_month
ORDER BY
  sale_month;
```

```python
%sql
SELECT
  DATE_FORMAT(`date_of_transfer`, 'yyyy-MM') AS `month`,
  `property_type`,
  COUNT(`transaction_id`) AS `transaction_count`
FROM
  `caio_moreno`.`uk_house_advisor_bot`.`silver_pp_2025`
WHERE
  YEAR(`date_of_transfer`) = 2025
GROUP BY
  DATE_FORMAT(`date_of_transfer`, 'yyyy-MM'),
  `property_type`
ORDER BY
  `month`,
  `transaction_count` DESC
```

```python
%sql
SELECT
  CONCAT('Q', QUARTER(`date_of_transfer`)) AS `quarter`,
  COUNT(`transaction_id`) AS `transaction_count`
FROM
  `caio_moreno`.`uk_house_advisor_bot`.`silver_pp_2025`
WHERE
  YEAR(`date_of_transfer`) = 2025
GROUP BY
  QUARTER(`date_of_transfer`)
ORDER BY
  `quarter`
```

```python
%sql
SELECT
  CONCAT('Q', QUARTER(`date_of_transfer`)) AS `quarter`,
  AVG(`price`) AS `average_price`
FROM
  `caio_moreno`.`uk_house_advisor_bot`.`silver_pp_2025`
WHERE
  YEAR(`date_of_transfer`) = 2025
GROUP BY
  QUARTER(`date_of_transfer`)
ORDER BY
  `quarter`
```
